#### Was ist Statistik?

In der Mathematik setzt sich das Gebiet der **Stochastik** aus der **Wahrscheinlichkeitstheorie** und der **Statistik** zu sammen.  Das Untergebiet der Statistik lässt sich in drei Bereiche einteilen:
- **Deskriptive Statistik**: reine Beschreibung einer Stichprobe durch grafische Darstellungen (z. B. Histogramme der Messwerte) und Kennzahlen (z. B. Mittelwert und Streuung)
- **Mathematische Statistik** (auch: Inferenzstatistik, statistische Inferenz, engl.: inference): Schließen von Stichprobe auf Eigenschaften der Grundgesamtheit. Eingesetzte Methoden sind z. B. die Schätzung von Parametern und Hypothesentests. Beispiel: Aus der Messung der Lebendauerverteilung eines Radionuklids wird dessen Halbwertszeit als Parameter eines physikalischen Modells des radioaktiven Zerfalls bestimmt.
- **Explorative Statistik**: Untersuchung von Daten, für die wenig Wissen über Zusammenhänge vorliegt, und Bildung neuer Hypothesen aus Daten (z. B. im Data Mining). Beispiel: Anhand der Verkaufszahlen eines Webshops wird ermittelt, and welchen weiteren Produkten Käufer:innen roter Schuhe Interesse haben könnten.


#### Physik und Statistik

In der Physik gibt es eine Reihe von Berührungspunkten mit der Statistik. Diese ergeben sich zum einen aus den physikalischen Prozessen selbst und zum anderen aus dem Vorgang des Messens in der Experimentalphysik. 

Fragen nach der **Vorhersagbarkeit** von Prozessen stellen sich bereits in der **klassischen Physik**, in der alle Gleichungen zur Beschreibung physikalischer Gesetzmäßigkeiten deterministisch sind. Damit sollte jeder klassische Prozess im Prinzip exakt vorhersagbar sein, z. B. die Bewegung eines physikalisches Pendel oder Planetenbahnen. Es sind aber auch Prozesse bekannt, die extrem abhängig von den Anfangsbedingungen sind und ggf. chaotisches Verhalten zeigen, was ihre Vorhersagbarkeit limitiert. Beispiele dafür sind das Dreikörperproblem der Gravitation oder das Doppelpendel. Ein anderes Extrem der klassischen Physik sind Vielkörpersysteme, für die eine Beschreibung über statistische Methoden geeigneter sind als eine Behandlung jedes einzelnen Körpers. Dazu gehören z. B. Gase: deren Temperatur lässt sich über eine statistische Eigenschaft der Gasmoleküle beschreiben (mittlere kinetische Energie), ohne die Flugbahn jedes einzelnen Moleküls zu kennen. 

In der **Quantenmechanmik** kommt hinzu, dass die Grundgleichungen zwar weiterhin deterministisch sind (z. B. die Schrödingergleichung zur Beschreibung der quantenmechanischen Wellenfunktion), über beobachtbare physikalischen Größen ("Observable") nur **Wahrscheinlichkeitsaussagen** machen können. In der gebräuchlichen Kopenhagener Deutung der Quantenmechanik entspricht das Quadrat der Wellenfunktion einer **Wahrscheinlichkeitsdichte**. Das Pendant zum klassischen Aufenthaltsort eines Objekts wird der **Erwartungswert** einer statistischen Aufenthaltswahrscheinlichkeit.

Die Physik lebt vom **Wechselspiel zwischen Experiment und Theorie**. In physikalischen Experimenten werden Messdaten gesammelt. Aus statistischer Sicht sind diese Messdaten eine **Stichprobe** (engl.: sample) aus der Grundgesamtheit aller möglichen Messergebnisse. Diese Stichprobe wird nun mit der Erwartung aus einem theoretischen Modell verglichen. Oft besitzen theoretische Modelle freie Parameter (z. B. der Widerstand $R$ im ohmschen Gesetz $U=RI$) die aus dem Vergleich des Modells mit den Daten bestimmt werden können. Mit dem Vergleich wird der **wahre Wert** dieser Parameter **geschätzt**. Das Wort *Schätzung* wird an dieser Stelle in seiner Bedeutung in der Fachsprache der Statistik verwerden: in der Statistik ist eine Schätzung ein systematisches Verfahren um Informationen über unbekannte Parameter eines Modells zu erhalten. Beachten Sie, dass dies nicht der Verwendung dieses Begriffs in der Alltagssprache entspricht, in der "schätzen" in etwa mit "ohne exaktes Messen näherungsweise bestimmen" übersetzt werden kann. In der Parameterschätzung ist heute der Computer das zentrale Werkzeug. Es kann gezeigt werden, dass im Grenzfall sehr großer Stichproben die Verteilung der geschätzten Parameter immer einer Normalverteilung folgt ("zentraler Grenzwertsatz", später). Aus diesem Grund hat die Normalverteilung in der Statistik eine besondere Bedeutung.

Die Schätzung ist immer mit einer **Unsicherheit** verbunden. In der Physik ist es üblich, diese Unsicherheit in eine **statistische**, eine **systematische** und ggf. eine **theoretische** Komponente zu zerlegen. Mit einer größeren Stichprobe lässt sich die statistische Unsicherheit der Schätzung reduzieren, womit sich die **Präzision** der Messung erhöht. Eine Reduktion der systematischen Unsicherheit lässt sich z. B. durch eine bessere Kalibrierung des Messgeräts oder ein besser ausgelegtes Experiment erreichen, was die **Richtigkeit** (Abstand vom wahren Wert) der Messung erhöht. Zusätzlich kann auch das theoretische Modell noch eine Unsicherheit besitzen, beispielsweise dadurch, dass darin Größen nur näherungsweise ausgerechnet werden konnten. Früher wurde an dieser Stelle oft der Begriff **Fehler** (engl.: error) anstatt von **Unsicherheit** verwendet. In diesem Kurs sind mit dem Begriff Fehler "echte" Fehler gemeint, die im Experiment natürlich auch passieren können (z. B. Kabel defekt), die aber vor der eigentlichen Messung entdeckt und korrigiert werden sollten.

## Modell einer Messung
Mit dem obigen Begriff von Messen und Unsicherheit wird nun ein Modell einer Messung entwickelt, das sich besonders gut zum Erlernen der grundlegenden Methoden der Datenauswertung eignet und mit dem gleichzeitig besonders einfach eine Messung simuliert werden kann, indem simulierte (auch: "synthetische") Daten mit Unsicherheiten erzeugt werden. Das Modell besagt, dass ein gemessenen Wert $m$ einer Größe sich ergibt aus der Summe aus dem wahren Wert $w$ und einem zufälligen Beitrag $z$ aufgrund der Unsicherheiten:
$$ m = w + z $$
Die Unsicherheiten teilen sich dabei auf in die drei Anteile
$$ z = z_\mathsf{stat} + z_\mathsf{syst} + z_\mathsf{theo}. $$
Dabei behinhalten die Anteile:
- $z_\mathbf{stat}$: statistische Unsicherheiten aufgrund zufälliger Fluktuationen des Messwerts ("Rauschen")
- $z_\mathbf{syst}$: systematische Unsicherheiten aufgrund der Richtigkeit von Messinstrument und Messmethode
- $z_\mathbf{theo}$: theoretische Unsicherheiten, d. h. Unsicherheiten im wahren Wert durch unterschiedliche Modellvorhersagen

Zur Simulation der Messung greift das Modell auf das "Würfeln" bzw. "Ziehen" des zufälligen Beitrags $z$ mithilfe von (Pseudo-)Zufallszahlen zurück. Die grundlegende Idee dabei ist, mehrfach hintereinander den wahren Wert um einen anderen zufälligen Betrag zu "verschmieren". Jeder dieser "verschmierten" Werte von $m$ steht dann für eine Messung. Somit entsteht ein sythetischer Datensatz, der die wesentlichen Eigenschaften einer Messung simuliert.

Ein einfaches Beispiel ist die Messung der Kreisfrequenz der harmonischen Schwingung eines Federpendels. Der wahre Wert ist (z. B. aus Masse und Federkonstante bekannt), z. B. $\omega_0 = 1\,\mathsf{s}^{-1}$. Die Zeitmessung ist aufgrund der verwendeten Stoppuhr nicht genauer als 0.1 Sekunden. Dies wird berücksichtigt, indem Zufallszahlen (siehe unten) mit einer Streuung von 0.1 (in der Einheit $\mathsf{s}^{-1}$) aus einer Normalverteilung gezogen und zum wahren Wert addiert werden.

```python
""" Simulation of measurement using a computer """

import numpy as np

# number of measurements
n = 10

# NumPy array of true values
w = 1.0 * np.ones( n )

# NumPy array of random contribution, simulated as 
# Gaussian random numbers with a mean value of 0 and a width of 0.1 s^-1
rng = np.random.default_rng( 42 )
z = rng.normal( 0, 0.1, n )

# print results
for i in range( n ):
	print( 'w = {:.2f}   m = w + z = {:.2f}'.format( w[i], w[i]+z[i] ) ) 
```

>**Exkurs: Zufall aus dem Computer**
>
> Wie schon kurz im Kurs "Programmieren und Algorithmen" diskutiert, sind **Zufallszahlen** ein wichtiges Werkzeug für eine Klasse von numerischen Techniken zur Berechnung von Wahrscheinlichkeiten. In der numerischen Mathematik werden diese Techniken zur Integration, Optimierung und Faltung verwendet und in der angewandten Statistik für Wahrscheinlichkeitsdichten und die Fortpflanzung von Unsicherheiten (früher: "Fehlerfortpflanzung"). Außerdem können stochastische Prozesse damit simuliert werden, z. B. in Vielteilchensystemen. 
>
> "Echte" Zufallszahlen können nur in echt zufälligen physikalischen Prozessen erzeugt werden, z. B. Rauschen oder radioaktive Zerfälle. Das macht es schwierig sie in großen Mengen effizient zu erzeugen. Viel effizienter in großen Mengen am Computer erzeugbar sind **Pseudozufallszahlen**, eine deterministische Sequenz von Zahlen, die praktisch nicht von echten Zufallszahlen unterschieden werden können. 
>
> Es gibt einige effiziente Algorithmen, um lange Sequenzen von (Milliarden von) Pseudozufallszahlen am Computer zu erzeugen und die in modernen **Zufallszahlengeneratoren** implementiert sind. Zur Illustration des Prinzips wird hier ein einfacher multiplikativer linear kongruenter Generator (engl.: linear congruential generator, LCG) verwendet, dessen Zufallszahlen allerdings nicht die Qualität besitzen, die in derzeitigen Anwendungen benötigt wird. Die Zufälligkeit kommt hierbei durch die Modulo-Operation, also das Dividieren zweier ganzer Zahlen mit Rest. Die Sequenz beginnt mit einem Startwert (engl.: seed) $x_0$, außerdem wird ein Divisor $m$ gewählt. Danach wird die nächste Zufallszahl über die Vorschrift
>$$
>x_{i} = (ax_{i-1} + c)\mod m, \quad i=1,2,3,\dots
>$$
>erzeugt. Um eine Zufallszahl zwischen 0 und 1 zu erhalten, dividieren wir dann noch durch $m$:
>$$
> r_i = x_i/m \in [0,1[ 
>$$
>Diese Algorithmus kann wie folgt implementiert werden:

```python
"""
simple implementation of a linear congruential generators (LCG)
T. Ferber, April 2022, 
further simplified and translated by U. Husemann, April 2024
"""

# x_i =(a * x_i-1 + c) mod m
# return value: r_i =x_i/m, with r_i in [0,1[
# Important: good initial values (m: prime number, c = 0)
def lcg( seed, m = 2**31-1, a = 1140671485, c = 0 ):
    next  = (a*seed + c) % m
    return next/m, next

# number of random numbers to be generated
n = 10

# seed: from now on the sequence is deterministic
seed = 1

for i in range( n ):
    rnd, seed = lcg(seed)
    print(i, rnd)
```

>In NumPy verwenden wir das Paket [`np.random`](https://numpy.org/doc/stable/reference/random/index.html), das auf Algorithmen mit besseren Eigenschaften als LCGs zurückgreift (derzeit voreingestellt: [PCG-64](https://numpy.org/doc/stable/reference/random/bit_generators/pcg64.html#numpy.random.PCG64)) und neben gleichverteilten Zufallszahlen auch solche mit vielen anderen statistischen Verteilungen erzeugen kann, z. B. normalverteilte oder exponentiell verteilte Zufallszahlen.

> **It's called Monte Carlo, because you're playing on someone else's money...** (zitiert nach B. Jacobson, Berkeley)

> Weil die hier beschriebenen numerischen Techniken das "Würfeln" von Werten mittels (Pseudo-)Zufallszahlen beinhalten, werden sie in Anlehnung an das bekannte Casino im Fürstentum Monaco **Monte-Carlo-Methoden** genannt. ## Wahrscheinlichkeitstheorie

Auch der Zufall folgt genauen mathematischen Regeln, die in der **Wahrscheinlichkeitstheorie**, einem Teilgebiet der Stochastik, beschrieben werden. Zunächst werden einige grundlegenden Begriffe eingeführt, die in der Stochastik verwendet werden:

> Als **Zufallsexperiment** (engl.: trial) wird ein Versuch bezeichnet, der unter genau festgelegten Bedingungen einen zufälligen Ausgang besitzt. Beispiele dafür sind das Würfeln von Zahlen, das Messen einer physikalischen Größe oder die Befragung einer Person auf der Straße nach ihrem Wahlverhalten.

> Unter **Zufallsereignis** oder kurz Ereignis (engl.: event) verstehen wir das Ergebnis eines Zufallsexperiment. In den obigen Beispielen wäre das eine "6" zu würfeln, eine Zeitspanne von 0,321 Sekunden zu messen und dass sich die Person als Wählerin der Partei XY herausstellt.

> Die **Wahrscheinlichkeit** (engl.: probability) für das Eintreten eines Ereignisses $A$ wird im folgenden mit $P(A)$ bezeichnet. Die **bedingte Wahrscheinlichkeit** (engl.: conditional probability) für das Eintreten von $A$, gegeben, dass Ereignis $B$ ebenfalls eintritt, wird mit $P(A|B)$ bezeichnet. Zwei Ereginisse $A$ und $B$ werden als **unabhängig** (engl.: independent) bezeichnet, wenn für ihre Schnittmenge (engl.: intersection) gilt $P(A\cap B)= P(A)\cdot P(B)$; in diesem Fall multiplizieren sich also ihre Wahrscheinlichkeiten.

Bereits für den zentralen Begriff der Wahrscheinlichkeit gibt in der Wahrscheinlichkeitstheorie mehrere mögliche Definitionen: die Komogorov-Axiome, den frequentistischen und den bayesschen Wahrscheinlichkeitsbegriff.

### Kolmogorov-Axiome

> Die **Kolmogorov-Axiomen** (Andrei N. Kolmogorov, 1933) erlauben eine abstrakte mathematische Definition der Wahrscheinlichkeit:
> Sei $S$ ein (abstrakter) Wahrscheinlichkeitsraum, d. h. eine nichtleere abzählbare Menge von Ereignissen. Für jede Untermenge von Ereignissen $A$ gibt es dann $P(A) \in \mathbb{R}$, so dass gilt:
> 1. Wahrscheinlichkeiten sind positiv (inklusive 0): $P(a) \geq 0$ für alle $A \in S$
> 1. Für paarweise disjunkte Ereignisse sind Wahrscheinlichkeiten additiv: für $A_i \cap A_j = \emptyset$ für alle $i\neq j$ gilt $P(\cup_i A_i) = \sum_i P(A_i)$
> 1. Wahrscheinlichkeiten sind normiert (auf eine Gesamtwahrscheinlichkeit von 1): $P(S)=1$.

Aus den Kolmogorov-Axiomen lassen sich einige **Folgerungen** für die Eigenschaften von Wahrscheinlichkeiten ableiten (Beweise bei Bedarf nachrechnen oder nachlesen):
1. $P(A) \in [0,1]$ und für das Komplement $\bar{A}$: $P(\bar{A}) = 1-P(A)$
1. $P(\emptyset) = 0$
1. $P(A\cup\bar{A}) = 1$
1. $P(A\cap\bar{A}) = 0$
1. $P(A\cup B) = P(A) + P(B) - P(A\cap B)$
1. $P(A\cap B) = P(A)\cdot P(B|A) = P(B)\cdot P(A|B)$
1. $P(B\setminus A) = P(B) - P(A\cap B)$
1. Aus $A\subseteq B$ folgt $P(A) \leq P(B)$

Symbolerklärung:
- $\subset$, $\subseteq$: Untermenge, Teilmenge (engl.: subset)
- $\emptyset$: leere Menge (engl.: empty set)
- $\cup$: Vereinigungsmenge (engl.: union)
- $\cap$: Schnittmenge (engl.: intersection)
- $\setminus$: "ohne" (relatives Komplement, engl.: set difference)

Mit Folgerung 6 ergibt sich eine besonders anschauliche Schreibweise für **bedingte Wahrscheinlichkeiten**:
$$
P(A|B) = \frac{P(A\cap B)}{P(B)} \textsf{ und }
P(B|A) = \frac{P(A\cap B)}{P(A)}
$$
Damit gilt für **unabhängige** Eereignisse $A$ und $B$: $P(A|B) = P(A)$ und $P(B|A)=P(A)$. Diese Zusammenhänge lassen sich grafisch in Venn-Diagrammen (John Venn, 1881) darstellen:

### Frequentistischer Wahrscheinlichkeitsbegriff

Der (klassische) **frequentistische Wahrscheinlichkeitsbegriff** definiert Wahrscheinlichkeiten über die relative Häufigkeit (Frequenz) von Zufallsereignissen. In diesem Kurs werden wir fast ausschließlich auf diesen Wahrscheinlichkeitsbegriff zurückgreifen. Er ist gut geeignet, wenn Messungen häufig wiederholbar sind, z. B. für einen Münzwurf, aber problematisch für singuläre Ereignisse, z. B. das Wetter von morgen oder den Urknall. 

>Im frequentistischen Ansatz ist die Wahrscheinlichkeit definiert als Limes der relativen Häufigkeit, in einer Messreihe mit einer unendliche Zahl von Messungen in $N$ Messungen $k$ mal das Ereignis $A$ zu beobachten:
>$$
P(A) = \lim_{N\to\infty}\frac{k}{N}
>$$

Aus dieser Definition ist sehr gut zu erkenne, dass der Wertebereich für $P(A)$ zwischen 0 und 1 liegt - in Übereinstimmung mit den Kolmogorov-Axiomen. 

In der frequentistischen Statistik existiert somit ein **"objektiver" Wahrscheinlichkeitsbegriff**: für eine Größe (z. B. einen Parameter eines physikalischen Modells) existiert ein "wahrer Wert". Dieser ist unbekannt, wird aber im Limes unendlich vieler Stichproben erreicht. Wenn in der frequentistischen Statistik Größen und deren Unsicherheit aus einer Messung geschätzt werden, ist ein typisches Resultat ein **Konfidenzintervall** (engl.: confidence interval), über das dann Aussagen getroffen werden wie "wenn diese Messung unendlich oft wiederholt und jeweils daraus ein Konfidenzintervall berechnet werden würde, läge der wahre Wert in 68% aller Fälle in diesem Intervall". Es wird hier also eine Wahrscheinlichkeitsaussage über das Konfidenzintervall gemacht, nicht über den wahren Wert. Dies steht im Gegensatz zum [bayesschen Wahrscheinlichkeitsbegriff](#bayes), siehe weiter unten.



Zufallsexperimente können mit Monte-Carlo-Verfahren einfach am Computer simuliert werden. Das Ergebnis eines fairen Münzwurfs kann entweder das Ereignis "Kopf" (engl.: obverse, heads) oder "Zahl" (engl.: reverse, tails) sein, beide mit 50% Wahrscheinlichkeit: $P(\mathsf{heads})=P(\mathsf{tails})=0.5$. Nach $N$ Würfen ist die relative Häufigkeit für "Kopf" gegen durch die Zahl der Ereignisse "Kopf" $N_h$ normiert auf $N$:
$$
h_N = \frac{N_h}{N}
$$
Im Limes unendlich vieler Würfe nähert sich $h_N$ der Wahrscheinlichkeit $P(\mathsf{head})$ an: $P(\mathsf{head})=\lim_{N\to\infty} h_N$. Später wird gezeigt, dass die Unsicherheit von $h_N$ mit $1/\sqrt{N}$ abfällt.


```python
''' example to produce random numbers (0/1) 
    as obtained by throwing a coin
    animated fraction of heads vs. number of throws
    
.. author:: Guenter Quast <g.quast@kit.edu> (a few tweaks by Ulrich Husemann)
'''

%matplotlib ipympl

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anim
import sys

nthrow = 499    # how often to throw

#create a figure
fig = plt.figure(figsize=(7.5,7.5))
ax  = fig.add_subplot(1,1,1)
ax.grid(True)
ax.set_xlabel('$N$ (number of trials)', size='x-large')
ax.set_ylabel('$h_{N}$ = $N_h / N$', size='x-large')
x = np.linspace(1,nthrow+1,nthrow)

# plot expectation ...
ax.plot(x,0.5*np.ones(nthrow),'r--',lw=2)
# ... and error band
#ax.plot(x,0.5+0.5/np.sqrt(x),'b--',lw=2)
#ax.plot(x,0.5-0.5/np.sqrt(x),'b--',lw=2)
ax.fill_between(x,0.5-0.5/np.sqrt(x),0.5+0.5/np.sqrt(x),color='lightblue',alpha=0.5)
txt1 = ax.text(0.1, 0.9, ' ', transform=ax.transAxes,size='xx-large',
  bbox=dict(facecolor='silver', edgecolor='r', boxstyle='circle', pad=0.5))
txt2 = ax.text(0.81, 0.93, ' ', transform=ax.transAxes,size='x-large',
                    backgroundcolor='white')

#
# throw the coin and plot N_head over (number of trials) 
Nh=0
N=0
rng = np.random.default_rng( 42 )

def animate(i):
  global N, Nh, r, x, rng
  N+=1
  if rng.random() >= 0.5:
    Nh+=1
    t=' O '
  else:
    t='-1-'
  r=float(Nh)/float(N)
  # plot result 
  graph, = ax.plot(float(N), r, 'g.')
  txt1.set_text(t)
  txt2.set_text('$N$ = %i' %(N))
  return graph, txt1, txt2
#
# show as animated (=updating) graph
print(('\n*==* script ' + sys.argv[0]+' executing'))
ani=anim.FuncAnimation(fig, animate, nthrow, interval=10, blit=False,
                         init_func=None, fargs=None, repeat=False)
plt.show()


```

### Bayesscher Wahrscheinlichkeitsbegriff
Der **bayessche Wahrscheinlichkeitsbegriff**, benannt nach [Thomas Bayes (1701-1761)](https://de.wikipedia.org/wiki/Thomas_Bayes) (ausgesprochen: ˈbɛɪ̯z), definiert Wahrscheinlichkeiten über den Grad der persönlichen Überzeugung (engl.: degree of belief), dass ein Ereignis eintritt. Dieser Ansatz einer "subjektiven" Wahrscheinlichkeit klingt zunächst wenig wissenschaftlich, erfüllt aber – wie der [frequentistische Ansatz](#freq)– die Kolmogorov-Axiome. Er hat den zusätzlichen Vorteil gegenüber dem frequentistischen Wahrscheinlichkeitsbegriff, dass er auch für Zufallsexperimente und -ereignisse anwendbar ist, die sie nicht wiederholen lassen (z. B. das Wetter morgen).

In der bayesschen Statistik gibt es das Konzept eines (festen) wahren Werts nicht. An dessen Stelle tritt eine (bedingte) Wahrscheinlichkeitsverteilung $P(\mathsf{truth}|\mathsf{data})$. Eine Stärke der bayesschen Statistik ist die Möglichkeit, auf Vorwissen (engl.: prior knowledge) über die Daten zurückzugreifen, z. B. aus einer vorherigen Messung. Dazu wird der **Satz von Bayes** verwendet, der sich aus den Kolmogorov-Axiomen und bedingten Wahrscheinlichkeiten ableiten lässt:
>$$
P(\mathsf{truth}|\mathsf{data}) = \frac{P(\mathsf{data}|\mathsf{truth})\cdot P(\mathsf{truth})}{P(\mathsf{data})}
>$$

Ein bayessches Konfidenzintervall (engl.: credible interval) hat eine völlig andere Aussage als ein frequentistisches Intervall, z. B. "der Mittelwert der Messungen liegt mit 68% Wahrscheinlichkeit im Intervall". Dies ist eine Aussage über den Mittelwert, nicht über das Intervall.

>**Satz von Bayes: Herleitung und Interpretation**
>
>Zur Herleitung des Satzes von Bayes vergleichen wir die bedingten Wahrscheinlichkeiten $P(A|B)$ und $P$(B|A)$:
>$$
\begin{split}
P(A|B) &= \frac{P(A\cap B)}{P(B)},\\
P(B|A) &= \frac{P(B\cap A)}{P(A)}.
\end{split}
>$$
>Wir beachten nun, dass gilt $P(B\cap A)=P(A\cap B)$. Damit können wir z. B. die zweite Gleichung nach dieser Größe auflösen und in die erste einsetzen:
>$$
\begin{split}
 \quad P(A \cap B) &= P (B\cap A) = P(B|A) \cdot P(A)\\
\to\quad P(A|B) &= \frac{P(A\cap B)}{P(B)} = \frac{ P(B|A) \cdot P(A)}{P(B)}.
\end{split}
>$$
>Für $A = \mathsf{truth}$ und $B=\mathsf{data}$ ist dies der Satz von Bayes:
>$$
P(\mathsf{truth}|\mathsf{data}) = \frac{P(\mathsf{data}|\mathsf{truth})\cdot P(\mathsf{truth})}{P(\mathsf{data})}
>$$
>
>Der Satz von Bayes erlaubt also **Wahrscheinlichkeitsaussagen über den wahren Wert** einer oder mehrerer Größen, z. B. physikalischer Observablen. Er beschreibt, wie die Wahrscheinlichkeit eines wahren Wertes für diese Größen berechnet werden kann, wenn die Messdaten bekannt sind. Das ist die Größe auf der linken Seite der Gleichung, $P(\mathsf{truth}|\mathsf{data})$. Diese Wahrscheinlichkeit darf *nicht* mit der Wahrscheinlichkeit gleichgesetzt werden, die Messdaten zu beobachten, wenn die Größen einen bestimmten wahren Wert besitzen, $P(\mathsf{data}|\mathsf{truth})$. Es muss außerdem die *a-priori-Wahrscheinlichkeit* des wahren Werts $P(\mathsf{truth})$ berücksichtigt werden. Diese Größen haben folgende Interpretation:
>- $P(\mathsf{data}|\mathsf{truth})$ wird aus den Messdaten durch statische Inferenz gewonnen, z. B. mit der Maximum-Likelihood-Methode (siehe Ausblick in Kapitel 4).
>- $P(\mathsf{truth})$ ergibt sich durch Vorwissen, z. B. aus früheren Messungen, das durch die Messdaten sozusagen ein Update bekommt. 
>- $P(\mathsf{data})$ ist ein zusätzlicher Normierungsfaktor, der oft als "Evidenz" bezeichnet wird. Betrachten wir die wahren Werte der Größen einzeln ($\mathsf{truth}_j$), dann kann er mit dem Satz der totalen Wahrscheinlichkeit so umgeschrieben werden, dass er sich aus den Messdaten und dem Vorwissen über die wahren Werte berechnen lässt zu:
>$$
P(\mathsf{data}) = \sum_j P(\mathsf{truth}_j \cap \mathsf{data}) = \sum_j P(\mathsf{data}|\mathsf{truth}_j) \cdot P(\mathsf{truth}_j).
>$$

## Häufigkeit und Wahrscheinlichkeitsverteilungen

### Zufallsvariable
Im folgenden wird der Begriff **Zufallsvariable** (engl.: random variable) häufig verwendet. Formal definiert ist dieser Begriff wie folgt:

> Als **Zufallsvariable** wird eine vom Zufall abhängige (hier oft reelle) Funktion $X$ bezeichnet, die einen Wahrscheinlichkeitsraum $S$ auf zufällige Ergebnisse $x$ im Messraum $E$ abbildet. Damit ordnet sie jedem möglichen Ereignis in $S$ eine Größe (z. B. eine reelle Zahl) zu.

Die Kolmogorov-Axiome und ihre Folgerungen zur Beschreibung von Wahrscheinlichkeiten "leben" in einem abstrakten  Wahrscheinlichkeitsraum (engl.: probability space) aller möglichen Messergebnisse. Im Wahrscheinlichkeitsraum gibt es eine Wahrscheinlichkeit für ein Messergebnis $x$. Im Messraum $E$ (engl.: measurable space) "leben" die Resultate einer Messung. In der Praxis wird diese formale Unterscheidung oft nicht gemacht und $x_i$ bzw. $x$ direkt als Zufallsvariable bezeichnet. In diesem Kurs werden wir in der Regel von Verteilungen sprechen, aus denen Zufallsvariablen gezogen bzw. gemessen werden.

Das folgende Codebeispiel demonstriert diese Abbildung vom Wahrscheinlichkeitsraum in den Messraum. Im Wahrscheinlichkeitsraum ist die Wahrscheinlichkeit für ein Messergebnis im Beispiel gegeben durch eine Normalverteilung $N(x;\mu,\sigma)$ mit festen Werten der Parametern $\mu$ und $\sigma$ (rote Kurve). Ergebnisse im Messraum werden durch 20 zufällige Werte von $x$ (blaue Striche) symbolisiert. Diese werden per Zufallszahlengenerator gewürfelt nach der Wahrscheinlichkeitsdichte $N(x;\mu,\sigma)$. 

```python
""""
illustration of a random variable as a mapping 
from probability space to measurement space
"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# probability space: true parameters of normal distribution known
true_mu, true_sigma = 0., 1.
 
rng = np.random.default_rng( 42 )
data = rng.normal( true_mu, true_sigma, size=20 )
mu, sig  = np.mean( data ), np.std( data, ddof=1 )
print( "Estimates of mu and sigma: %5.3f, %5.3f" % ( mu, sig ) )

# plot probability density from which measurements were drawn
x = np.linspace( -3, 3, 200 )
pdf = norm.pdf( x, loc=true_mu, scale=true_sigma )

# plot measurements as lines
fig,ax = plt.subplots()
ax.vlines( data, 0, 0.1, linestyles='-', lw=2)
ax.plot( x, pdf, 'r' )
ax.set_xlim( -3.3, 3.3 )
ax.set_ylim( 0, 0.5 )
ax.set_xlabel( "$x$" )
ax.set_ylabel( "Probability Density")

plt.show()
```

### Diskrete und kontinuierliche Ergebnisse
Zufallsvariablen können Ergebnisse von Messungen mit **diskreten** Werten $x_i$ (z. B. Kopf oder Zahl für einen Münzwurf) oder mit **kontinuierlichen** Werten $x$ (z. B. Längenmessung, Zeitmessung) beschreiben. 

Die Wahrscheinlichkeit für **diskrete** Ergebnisse wird mit einer **Wahrscheinlichkeitsfunktion** $P_X(x_i)$ (engl.: probability mass function, PMF) beschrieben, die die Wahrscheinlichkeit angeben, dass ein Ergebnis den diskreten Wert $x_i$ annimmt. Die PMF ist so normiert, dass die Summe über alle möglichen Wahrscheinlichkeiten den Wert 1 ergibt: $\sum_{E} P_X(x_i) = 1$.

Oft wird auch die Summe aller Wahrscheinlichkeiten vom kleinsten möglichen Wert bis zu einem Wert $x_0$ zur Beschreibung verwendet. Diese wird als **Verteilungsfunktion** (engl.: cumulative distribution function, CDF) und ist in einer Dimension definiert als
$$
F(x_0) = \sum\limits_{x_i \leq x_0} P_X(x_i).
$$

Als Beispiel für die PMF und CDF für Messungen mit diskrete Werten dient hier ein (ungezinkter) Würfel, bei dem der Wahrscheinlichkeitsraum, d. h. der Raum aller möglichen Ergebnisse, die Augenzahlen von 1 bis 6 beinhaltet. Jede dieser sechs Augenzahlen ist gleich wahrscheinlich. Mit der Normierung der PMF auf Eins folgt also $P_X(x_i) = 1/6$.

```python
""" 
Probability mass function and cumulative distribution function 
for discrete measurement results: example of dice
"""
import numpy as np
import matplotlib.pyplot as plt

# number of faces on dice
n = 6

# for probability: array with exactly one entry for each number
throws = np.arange( 1, n+1 )

# probability mass function
pmf = 1/n * np.ones( n )

# commulative distribution function
cdf = np.cumsum( pmf )

fig, ax = plt.subplots( 2, 1, figsize=([7,10]) )

# alternative representation using histograms
#ax[0].hist( throws, bins=np.linspace( 0.5, 6.5, 7 ), density=True )

# representation with bar charts
ax[0].bar( throws, pmf )
ax[0].set_xlabel( "Number")
ax[0].set_ylabel( "Probability Mass Function")

# alternative representation using histograms
#ax[1].hist( throws, bins=np.linspace( 0.5, 6.5, 7 ), density=True, cumulative=True )

# representation with bar charts
ax[1].bar( throws, cdf )
ax[1].set_xlabel( "Number")
ax[1].set_ylabel( "Cumulative Distribution Function")

plt.show()
```

Für **kontinuierliche** Ergebnisse, die jeden reellen Wert $x$ in einem Intervall im Wahrscheinlichkeitsraum einnehmen können, werden Wahrscheinlichkeiten durch **Wahrscheinlichkeitsdichtefunktionen** $f(x)$ (engl.: probability density function, PDF) beschrieben. Die PDF ist, wie ihr Namen schon sagt, selbst keine Wahrscheinlichkeit, sondern eine Wahrscheinlichkeitsdichte, die noch über einen Bereich von $x$ integriert werden muss um eine Wahrscheinlichkeit zu erhalten. Die (differenzielle) Wahrscheinlichkeit einen Wert im Intervall $[x_0,x_0+\mathsf{d}x]$ zu messen beträgt $f(x_0)\,\mathsf{d}x$. Die Wahrscheinlichtkeit exakt $x_0$ zu messen verschwindet dagegen. Die Normierung der PDF erfordert, dass das Integral über den gesamten Wahrscheinlichkeitsraum (in einer Dimension häufig $x\in\mathbb{R}$) den Wert 1 ergibt: $\int f(x)\,\mathsf{d}x = 1$. 

 Wie für die PMF einer diskreten Verteilung gibt es für die PDF einer kontinuierlichen Verteilung eine CDF, für $x\in\mathbb{R}$ definiert als
$$
F(x_0) = \int_{-\infty}^{x_0} f(x)\,\mathsf{d} x.
$$

Als Beispiel für die PDF und CDF von kontinuierlichen Verteilungen wird hier die **Normalverteilung** (auch: Gaußverteilungen) gezeigt, die im folgenden Abschnitt genauer definiert wird. In ihrer Standardform ist sie definiert als:
$$
\varphi(x) = \frac{1}{\sqrt{2\pi}}
\exp\left[ 
-\frac{x^2}{2}
\right].
$$

```python
"""PDF and CDF of normal (Gaussian) distribution: easily displayed using scipy.stats"""
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

#  x values within 0.1% and 99.9% percentiles
x = np.linspace( norm.ppf(0.001), norm.ppf(0.999), 200 )
pdf = norm.pdf( x )
cdf = norm.cdf( x )

fig, ax = plt.subplots( 2, 1, figsize=([7,10]) )

ax[0].plot( x, pdf )
ax[0].set_xlabel( "$x$")
ax[0].set_ylabel( "Probability Density Function" )

ax[1].plot( x, cdf )
ax[1].set_xlabel( "$x$")
ax[1].set_ylabel( "Cumulative Distribution Function" )

plt.show()
```

### Häufigkeitsverteilungen

Damit wird eine Messreihe simuliert, wie sie in einem Experiment hätte aufgezeichnet werden können. Im folgenden Beispiel werden $N=100$ Würfe mit einen Würfel mit sechs Seiten simuliert. In Histogrammen wird die Häufigkeitsverteilung und die kumulative Häufigkeitsverteilung (normiert auf eins) grafisch dargestellt (für eine diskrete Verteilung hätten auch Balkendiagramme verwendet werden könnten).


```python
"""simulation of playing dice based on random numbers"""
import numpy as np
import matplotlib.pyplot as plt

# initialize random number generator with seed
rng = np.random.default_rng( 42 )

# N random throws of six-face die
N = 100
throws = rng.integers( 1, 7, size=N )

fig, ax = plt.subplots( 2, 1, figsize=([7,10]) )

# plot histogram of PMF (could have also been a bar chart)
ax[0].hist( throws, bins=np.linspace( 0.5, 6.5, 7 ) )
ax[0].set_xlabel( "Number")
ax[0].set_ylabel( "Frequency")

# plot histogram of CDF (cumulative=True) normalized to unity (with density=True)
ax[1].hist( throws, bins=np.linspace( 0.5, 6.5, 7 ), density=True, cumulative=True )
ax[1].set_xlabel( "Number")
ax[1].set_ylabel( "Cumulative Frequency (Normalized)")

plt.show()
```

### Charakterisierung von Verteilungen
Es gibt eine Reihe von Möglichkeiten, statistische Verteilungen zu beschreiben. Die volle Information über die Verteilung steckt in der PMF bzw. PDF oder alternativ in der CDF der Verteilung. Einige wichtige Eigenschaften von Verteilungen sind auch ohne volle Kenntnis der PMF bzw. PDF (näherungsweise) beschreibbar, z. B. die Lokalisierung durch den Mittelwert. Der Mittelwert ist ein Beispiel für ein algebraisches **Moment** der Verteilung. Die **Schätzung** der Eigenschaften von Verteilungen aus **Daten** mit Methoden der mathematischen Statistik erlaubt es dann, auf die "wahre" Verteilung im Wahrscheinlichkeitsraum rückzuschließen. Folgende typische Merkmale sind zur Charakterisierung von Verteilungen geeignet:
- **Lokalisierung** der Verteilung (Frage: "Wo liegt die größe Wahrscheinlichkeit?”)
- **Streuung** um Lokalisierung der Verteilung (Frage: "Wie breit ist die Verteilung?”)
- **Form** der Verteilung (Frage: "Wie schief und/oder wie spitz ist die Verteilung?”)

Wichtige Kennzahlen zur Charakterisierung von Verteilungen sind deren **Quantile**. Sie beantworten die Frage, für welchen Wert $x_\alpha$ eine (kumulierte) Wahrscheinlichkeit $P(x_\alpha) = \alpha$ für die Verteilung erreicht ist. Dazu wird die CDF der Verteilung verwendet (wie oben für $x\in\mathbb{R}$), für diskrete Verteilungen
$$
F(x_\alpha) = \sum\limits_{x_i\leq x_\alpha} P_X(x_i) \overset{!}{=} \alpha
$$
und für kontinuierliche Verteilungen
$$
F(x_\alpha) = \int_{-\infty}^{x_\alpha} f(x)\,\mathsf{d} x \overset{!}{=} \alpha.
$$
Der Wert $x_\alpha$ wird dann als $\alpha$-Quantil der Verteilung bezeichnet. Zur Bestimmung von $x_\alpha$ muss obige Gleichung nach $x_\alpha$ aufgelöst werden. Mathematisch entspricht das dem Bilden der Umkehrfunktion der CDF:
$$
x_\alpha = F^{-1}(\alpha).
$$
Diese Umkehrfunktion lässt sich in einigen Fällen analytisch bilden, in anderen Fällen muss sie numerisch berechnet werden. 

Bekannte Beispiele für Quantile:
- Eine kontinuierliche Wahrscheinlichkeitsverteilung lässt sich immer "in der Mitte" unterteilen, so dass zwei Bereiche gleicher Wahrscheinlichkeit entstehen. Das so definierte 50%-Quantil $x_{0.5}$ heißt **Median**, es ist ein Maß für die **Lokalisierung** einer Verteilung.
- In vielen wissenschaftlichen Disziplinen ist es üblich, die Wahrscheinlichkeitsverteilung in vier Unterbereiche gleicher Wahrscheinlichkeit zu unterteilen. Neben dem Median werden dazu außerdem die 25%- und 75%-Quantile benötigt, die auch als erstes und drittes **Quartil** bezeichnet werden. Im Bereich zwischen dem ersten und dem dritten Quartil (engl.: inter-quartile range) liegt also eine Wahrscheinlichkeit von 50%. In dieser Vorlesung werden Sie Box-(Whisker)-Plots kennenlernen, die diese Quartile als Maß für die **Streuung** einer Verteilung um ihre Lokalisierung verwenden.
- In der Physik ist es dagegen üblich Quantile, die sich aus der Standardnormalverteilung (siehe oben) ergeben, als Maß für die Streuung zu verwenden. Der Mittelwert der Standardnormalverteilung ist 0, die Streuung um den Mittelwert ("Standardabweichung", wird weiter unten eingeführt) ist $\pm 1$. Die dazugehörigen Quantile entsprechen $\alpha \approx 0.159$ and $\alpha \approx 0.841$, sie decken also einer Wahrscheinlichkeit von ungefähr 68% (mit korrekter Rundung eigentlich 68.3%) ab. 

>**Bemerkung**: Es gibt eine Reihe von Verfahren zur Erzeugung von **Zufallszahlen mit beliebigen Verteilungen**, z. B. die [Inversionsmethode](https://de.wikipedia.org/wiki/Inversionsmethode) (bzw. Transformationsmethode), die wie Quantile auf der Umkehrfunktion der CDF beruht. Diese Verfahren werden z. B. in Wahlpflichtkurs "Datenanalyse" im fünften Fachsemester des Bachelorstudiengangs Physik am KIT behandelt.

```python
"""CDF and quantiles of normal (Gaussian) distribution: easily displayed using scipy.stats"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# x values within 0.1% and 99.9% percentiles
amin, amax = 0.001, 0.999
x = np.linspace( norm.ppf(amin), norm.ppf(amax), 200 )
cdf = norm.cdf( x )

# alpha values
alpha = np.linspace( amin, amax, 200 )
ppf = norm.ppf( alpha )

fig, ax = plt.subplots( 2, 1, figsize=([7,10]) )

ax[0].plot( x, cdf )
ax[0].set_xlabel( "$x$")
ax[0].set_ylabel( "Cumulative Distribution Function" )

ax[1].plot( alpha, ppf )
ax[1].set_xlabel( r"$\alpha$")
ax[1].set_ylabel( r"Quantiles $x_{\alpha}$" )

# add horizontal and vertical lines at median and plus/minus one standard deviation
quantile = np.array([-1,0,1])
alpha    = norm.cdf( quantile )
ax[1].hlines( quantile, amin, alpha, linestyles='dashed', colors='black')
ax[1].vlines( alpha, norm.ppf(amin), quantile, linestyles='dashed', colors='black')

# add text for values of alpha for one standard deviation
for a in alpha:
    ax[1].text( a+0.02, norm.ppf(amin), "%5.3f" % a  )

plt.show()
```

#### Lagemaß von Verteilungen
Es gibt mehrere Möglichkeiten, die **Lokalisierung** (auch: Lage) einer Verteilung zu beschreiben. Übliche Lokalisierungsparmeter sind:
- Der **Modus** (engl.: mode) entspricht dem wahrscheinlichsten Wert einer Verteilung, also dem Maximum der PDF oder PMF. 
- Der **Mittelwert** (engl.: mean) ist das gebräuchlichste Lagemaß in der Datenanalyse. Für PDFs und PMFs wird er über den Erwartungswert der Verteilung ausgerechnet. 
- Der **Median** als 50%-Quantil ist ebenfalls ein gängiges Lagemaß. Er kommt vor allem zum Einsatz in Verteilungen, die große Ausreißer (engl.: outliers) besitzen, denn er ist weniger anfällig für Verschiebungen aufgrund von Ausreißern als der Mittelwert. Beispiel: In einer Vermögensverteilung kann eine einzige Milliadärin den Mittelwert stark nach oben verschieben, der Median ist unabhängig vom Vermögenswert dieser Person und hängt nur davon ab, wie viele Personen mehr oder weniger Vermögen besitzen.

Für eine **symmetrische** Verteilung wie die Normalverteilung besitzen Modus, Mittelwert und Median denselben Wert, für asymmetrische Verteilungen wie z. B. die Log-Normalverteilung (PDF: $f(x) = 1/(\sqrt{2\pi} x)\exp\left[-\ln(x)^2/2\right]$) sind diese deutlich unterschiedlich, wie in folgendem Codebeispiel demonstriert:


```python
"""PDF of the log-normal distribution with different localization parameters"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import lognorm

x = np.linspace( 0, 5, 200 )

scale = 1.0
pdf = lognorm.pdf( x, s=scale )

localization = [ np.exp(-1), lognorm.median( s=scale ), lognorm.mean( s=scale ) ]
locname = [ "Mode", "Median", "Mean" ]
colors = [ "red", "green", "blue" ]

ymin, ymax = 0, 1.1*np.amax(pdf)

fig,ax = plt.subplots()
ax.plot( x, pdf )
ax.set_ylim( ymin, ymax )
ax.vlines( localization, ymin, ymax, colors=colors, linestyles="dashed" )
for l,n,c in zip(localization, locname, colors):
    ax.text( l+0.05, lognorm.pdf( l, s=scale )+0.01, n, color=c )

ax.set_xlabel( "$x$")
ax.set_ylabel( "Probability Density Function" )

plt.show()
```

#### Erwartungswert und Varianz

>Der **Erwartungswert** $\mathsf{E}[x]$ (engl.: expected value, expectation value) einer diskreten Verteilung ist definiert als
>$$ 
\mathsf{E}[x] \equiv \sum_{x_i\in E} x_i \cdot P_X(x_i).
>$$
>Für kontinuierliche Verteilungen mit $x\in\mathbb{R}$ lautet die Definition
>$$
\mathsf{E}[x] \equiv \langle x\rangle \equiv \mu \equiv  \int_{-\infty}^\infty x\cdot f(x)\, \mathsf{d} x.
>$$

In der Literatur gibt es für den Erwartungswerte eine Reihe unterschiedlicher Schreibweisen, die wichtigsten ($\mathsf{E}[x],\, \langle x\rangle,\, \mu$) sind in obiger Defintion zu sehen. Folgende Eigenschaften des Erwartungswerts werden im folgenden verwendet:
- $\mathsf{E}[x]$ ist ein **Operator**, dessen Anwendung auf eine Verteilung in einer **Zahl** resultiert (und nicht in einer Funktion). Damit ist der Erwartungswert eines Erwartungswert der Erwartungswert selbst: $\mathsf{E}[\mathsf{E}[x]]=\mathsf{E}[x]$.
- Die Bildung des Erwartungswerts ist eine **lineare** Operation. Daraus folgt, dass für den Erwartungswert einer Konstante $a$ gilt $\mathsf{E}[a]=a$ und dass $\mathsf{E}[ax_1+bx_2]=a\mathsf{E}[x_1]+b\mathsf{E}[x_2]$. Dies lässt sich durch einfaches Nachrechnen mit der Definition des Erwartungswertes zeigen.
- Der Erwartungswert ist im allgemeinen **nicht multiplikativ**: $\mathsf{E}[ x_1 \cdot x_2 ] \underset{\mathsf{i.\,allg.}}{\neq} \mathsf{E}[x_1]\cdot \mathsf{E}[x_2]$. Man kann zeigen, dass $\mathsf{E}[ x_1 \cdot x_2 ] = \mathsf{E}[x_1]\cdot \mathsf{E}[x_2]$ nur gilt, falls $x_1$ und $x_2$ unabhängig sind (Definition später).


>**Nebenrechnungen Erwartungswert**
>
>Die Linearität des Erwartungswerts ergibt sich aus seiner Definition, allerdings unter Vorgriff auf zweidimensionale Wahrscheinlichkeitsverteilungen, die später in Kapitel 2 diskutiert werden. Unter Verwendung der Normierung der PDF $f(x)$ auf Eins ergibt sich für den Erwartungswert einer Konstanten:
>$$
\mathsf{E}[a] = \int_{-\infty}^\infty a \cdot f(x)\, \mathsf{d}x = a \int_{-\infty}^\infty f(x)\, \mathsf{d}x = a
>$$
>und für den Erwartungswert einer Linearkombination $z = ax + by$
>$$
\begin{split}
\mathsf{E}[z] &= \int_{-\infty}^\infty z \cdot f(z)\,\mathsf{d}z \\
&= \int_{-\infty}^\infty \int_{-\infty}^\infty(ax + by) \cdot f(x,y) \, \mathsf{d}x\, \mathsf{d}y\\
&= a \int_{-\infty}^\infty \int_{-\infty}^\infty x \cdot f(x,y)\, \mathsf{d}x\, \mathsf{d}y +
b \int_{-\infty}^\infty \int_{-\infty}^\infty y \cdot f(x,y)\, \mathsf{d}x\, \mathsf{d}y.\\
\end{split}
>$$
>Wir führen jetzt im ersten Summanden die Integration über $y$ aus, so dass 
>$$
f_x(x) \equiv \int_{-\infty}^\infty f(x,y)\, \mathsf{d}y.
>$$
>Diese Größe werden wir später in diesem Kapitel noch als Rand-PDF (engl.: marginal PDF) kennenlernen. Analog integrieren wir im zweiten Summanden über $x$ und führen $f_y(y)$ ein. Damit ergibt sich für den Erwartungswert
>$$
\mathsf{E}[ax_1 + bx_2] = 
a \int_{-\infty}^\infty x\cdot f_x(x)\, \mathsf{d}x + b \int_{-\infty}^\infty y\cdot f_y(y)\, \mathsf{d}y = 
a\mathsf{E}[x] + b\mathsf{E}[y].
>$$
>
>Der Erwartungswert für das Produkt $z=xy$ zweier Zufallsvariablen ergibt sich mit einem analogen Ansatz.
>$$
\mathsf{E}[z] = \int_{-\infty}^\infty z \cdot f(z)\,\mathsf{d}z = \int_{-\infty}^\infty \int_{-\infty}^\infty xy \cdot f(x,y) \, \mathsf{d}x\, \mathsf{d}y.
>$$
>Dieses Doppelintegral über ein Produkt lässt sich, im Gegensatz zu dem über eine Summe oben, nicht in jedem Fall in zwei Teile zerlegen. Diese Zerlegung funktioniert nur, wenn $f(x,y)=f_x(x)\cdot f_y(y)$, wenn die gemeinsame PDF also das Produkt der beiden Rand-PDF ist. In diesem Fall bezeichnen wir $x$ und $y$ als unabhängig und nur dann können wir schreiben
>$$
\mathsf{E}[xy] = 
\left(\int_{-\infty}^\infty x\cdot f_x(x)\, \mathsf{d}x\right) \cdot
\left(\int_{-\infty}^\infty y\cdot f_y(y)\, \mathsf{d}y\right) = \mathsf{E}[x] \cdot \mathsf{E}[y].  
>$$

Die **Streuung** einer Verteilung um ihre Lokalisierung, manchmal auch als "Skalierungsparameter" bezeichnet, wird in der Regel durch deren **Varianz** charakterisiert.

>Die **Varianz** $\mathsf{V}[x]$ (engl.: variance) einer diskreten Verteilung ist definiert als
>$$ 
\mathsf{V}[x] \equiv  \sigma_x^2 \equiv \sum_{x_i\in E} (x_i - \mathsf{E}[x])^2 \cdot P_X(x_i).
>$$
> Die Definition für kontinuierliche Verteilungen mit $x\in\mathbb{R}$ lautet entsprechend:
>$$
\mathsf{V}[x] \equiv  \sigma_x^2 \equiv  \int_{-\infty}^\infty (x - \mathsf{E}[x])^2 \cdot f(x)\, \mathsf{d} x.
>$$

Folgende Eigenschaften der Varianz werden im folgenden relevant sein:
- Ein Vergleich mit der Definition des Erwartungswerts ergibt, dass sich Varianz auch schreiben lässt als $\mathsf{V}[x] = \mathsf{E}[ (x-\mathsf{E}[x])^2]$. Damit wird ihre Interpretation als Maß für die **Streuung** um die Lokalisierung klar: die Varianz ist der Erwartungswert der quadratischen Abweichung vom Mittelwert.
- Mit obiger Formulierung der Varianz und den Eigenschaften des Erwartungswerts lässt sich einfach nachrechnen, dass sich die Varianz wie folgt ausrechnen lässt (Beweis als Nebenrechnung unten): 
$$
\mathsf{V}[x] = \mathsf{E}[x^2] - \mathsf{E}[x]^2.
$$
- Obige Gleichung wird auch als **Verschiebungssatz** bezeichnet und ist mathematisch analog zum Satz von Steiner für Trägheitsmomente in der Mechanik. In praktischen Rechnungen wird dieser Art der Varianzberechung häufig verwendet, da $\mathsf{E}[x]$ vorher separat berechnet werden kann. Andererseits könnte die Varianz bei einer Berechnung mit Gleitkommazahlen am Computer numerisch instabil sein, wenn $\mathsf{E}[x^2]$ und $\mathsf{E}[x]^2$ ähnlich groß sind und sich in der Differenz (fast) auslöschen.
- Wenn die Größe $x$ eine Einheit $\mathsf{A}$ besitzt, dann ist die Einheit der Varianz dieselbe wie die von $x^2$, also $\mathsf{A^2}$.
- Oft ist es praktisch, ein Maß für die Streuung um die Lokalisierung einer Verteilung zu besitzen, das dieselbe Einheit hat wie $x$ selbst. Wie durch die Schreibweise $\mathsf{V}[x] \equiv  \sigma_x^2$ angedeutet, ist die Größe $\sigma_x$, die Wurzel der Varianz, dazu geeignet. Sie wird als **Standardabweichung** (engl.: standard deviation) bezeichnet.
- Eine Änderung der Varianz hat eine **Dehnung** oder **Stauchung** einer Verteilung zur Folge, sie ist allerdings invariant unter Verschiebung der Verteilung, denn sie ist relativ zum Erwartungswert definiert.


>**Nebenrechnung Varianz**:
>
>Mit der Definition der Varianz und den oben eingeführten Eigenschaften des Erwartungswerts lässt sich der Verschiebungssatz durch Nachrechnen beweisen:
>$$
\begin{split}
    \mathsf{V}[x] &= \mathsf{E}\left[ (x - \mathsf{E}[x])^2 \right]\\
    &=  \mathsf{E}\left[ x^2 - 2x\mathsf{E}[x] + \mathsf{E}[x]^2 \right]\\
    &\overset{\mathsf{linear}}{=} \mathsf{E}[x^2] 
    - 2\mathsf{E}\left[ x\underbrace{\mathsf{E}[x]}_{\mathsf{Zahl}}\right] 
    + \mathsf{E}\left[{\underbrace{\mathsf{E}[x]}_{\mathsf{Zahl}}}^2 \right ]\\
    &= \mathsf{E}[x^2] - 2\mathsf{E}[x]\cdot\mathsf{E}[x] + \mathsf{E}[x]^2\\
    &= \mathsf{E}[x^2] - \mathsf{E}[x]^2.\\
\end{split}
>$$

Für die **Normalverteilung** (auch: Gaußverteilung), gegeben durch 
$$
f(x;\mu,\sigma) \equiv N(x;\mu,\sigma) \equiv \frac{1}{\sqrt{2\pi}\sigma}
\exp\left[ 
-\frac{(x-\mu)^2}{2\sigma^2},
\right]
$$
lassen sich Erwartungswert und Varianz direkt aus den Parametern der Verteilung ablesen: 
$$
E[x] = \mu, \quad
V[x] = \sigma^2.
$$
Dies lässt sich durch Nachrechnen mit den Definitionen von Erwartungswert und Varianz zeigen. Die oben bereits verwendete Standardnormalverteilung ergibt sich aus der Normalverteilung für $\mu=0$ und $\sigma=1$. Verschiebung und Dehnung bzw. Stauchung der Normalverteilung werden mit folgendem Python-Code grafisch dargestellt, die Länge der gestrichelten Linien entspricht dabei der Standardabweichung.

```python
"""Normal (Gaussian) distributions with different expected value and variance"""

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# three different sets of parameters mu and sigma
param = np.array( [ [ 0., 1.] , [-1., 0.5], [ 0., 2. ] ] )

x = np.linspace( -5, 5, 200 )
pdf  = [ norm.pdf( x, loc=l, scale=s ) for l, s in param ]
ysig = [ norm.pdf( s, scale=s ) for l, s in param ]

ymin, ymax = 0, 1.1*np.amax(pdf)

for p, par in zip( pdf, param ):
	plot = plt.plot( x, p, label=r"$\mu = %3.1f,\,\sigma = %3.1f$" % (par[0],par[1])  )
	sigmin, sigmax = par[0]-par[1], par[0]+par[1]
	ysig = norm.pdf( sigmax, loc=par[0], scale=par[1] )
	plt.hlines( ysig, sigmin, sigmax, color=plot[-1].get_color(), linestyle="dashed" )
	
plt.ylim( ymin, ymax )
plt.xlabel( "$x$")
plt.ylabel( "Probabillity Density Distribution" )
plt.legend()

plt.show()
```

>$$

#### Schätzung von Erwartungswert und Varianz
Bisher haben wir in diesem Kurs Größen zur Charakterisierung von Verteilungen im Wahrscheinlichkeitsraum wie Erwartungswert und Varianz als mathematische Operatoren **definiert**. In der mathematischen Statistik wird der umgekehrte Weg verfolgt: es ist das Ziel, diese Größen mit möglichst hoher Präzision und Richtigkeit **aus Daten zu schätzen**. Dieses statistische Verfahren machen wir uns in der Experimentalphysik zu nutze, indem wir aus Messdaten die Parameter physikalischer Modelle (und deren Unsicherheit) schätzen. Zur Parameterschätzung aus Daten gibt es systematische Verfahren: Später in diesem Kurs wird die Methode der **kleinsten Quadrate** eingeführt. Die noch allgemeinere und leistungsfähigere **Maximum-Likelihood-Methode** lernen Sie kurz gegen Ende dieses Kurses und im Wahlpflichtkurs "Datenanalyse" kennen.

Der in vielerlei Hinsicht "beste" Schätzwert (engl.: estimator) für den Erwartungswert $\mathsf{E}[x]\equiv\mu$ einer Verteilung aus einer Stichprobe $\mathbf{x}=(x_1, x_2, \dots x_N)$ ist das **arithmetische Mittel** (engl.: arithmetic mean, oft kurz: mean) der Ergebnisse der Stichprobe: 
$$
\hat\mu \equiv \frac{1}{N} \sum\limits_{i=1}^{N} x_i.
$$
Das arithmetische Mittel ist eine Funktion der Daten, die Summe aller Ergebnisse der Stichprobe $x_i$ dividiert durch die Größe der Stichprobe $N$. In NumPy wird das arithmetische Mittel berechnet durch `muhat = np.mean(x)`.
Hier und im folgenden werden Schätzwerte für Größen immer mit einem Dach/Hut auf dem Symbol gekennzeichnet.

Soll die Varianz $\mathsf{V}[x]\equiv\sigma_x^2$ einer Verteilung geschätzt werden, muss zunächst der Erwartungswert über das arithmetische Mittel $\hat\mu$ geschätzt werden. Der Schätzwert für die Varianz beinhaltet dann die quadratische Abweichung jedes Ergebnisses zum arithmetischen Mittel $(x_i-\hat\mu)^2$ (auch: "Residuenquadrat") und eine geeignete Normierung. Es kann gezeigt werden, dass derselbe Normierungsfaktor wie beim arithmetischen Mittel, $1/N$, korrigiert werden muss, so dass er $1/(N-1)$ beträgt, damit die Varianz "unverzerrt" (engl.: unbiased) geschätzt wird. Der Korrekturfaktor dieser **Besselkorrektur** ist also $N/(N-1)$. Damit ergibt sich die **Stichprobenvarianz** (engl.: sample variance) als Schätzwert für die Varianz:
$$
\hat \sigma^2 = 
{\color{red}\frac{N}{N-1}} \frac{1}{N}
\sum\limits_{i=1}^{N} ( x_i - \hat\mu )^2 =
\frac{1}{N\color{red}{-1}} 
\sum\limits_{i=1}^{N} ( x_i - \hat\mu )^2.
$$
Die **Stichprobenstandardabweichung** $\hat\sigma$ ist die Wurzel der Stichprobenvarianz. In NumPy muss für die Berechnung der Stichprobenvarianz und -standardabweichung angegeben werden, ob die Besselkorrektur verwendet werden soll: mit `sig2hat = np.var( x, ddof=1 )` wird auf $N-1$ normiert.

> **Besselkorrektur** (Friedrich Wilhelm Bessel, 1784-1846)
> 
> Eine geeignete Normierung des Schätzwerts für die Varianz auf die Größe der Stichprobe zu finden erfordert eine Zusatzüberlegung. Mit derselben Normierung auf $1/N$ wie im arithmetischen Mittel stellt sich nämlich heraus, dass mit dieser Normierung für sehr kleine Stichproben die Varianz zu klein abgeschätzt werden würde und damit die Schätzung verzerren würde. 
>
>Für $N=1$ gibt es keine Streuung um den Mittelwert, die Varianz sollte eigentlich gar nicht definiert sein, denn es gibt gar kein unabhängiges Residuenquadrat $(x_i-\hat\mu)^2$ mit $i>1$. Für $N=2$ haben $x_1$ und $x_2$ denselben Wert von $(x_i-\hat\mu)^2$, denn das arithmetische Mittel liegt genau in der Mitte zwischen den Werten, es gibt also nur *ein* unabhängiges Residuenquadrat. Wenn jetzt mit $1/N=1/2$ normiert würde, wäre die Varianz um einen Faktor 2 zu klein abgeschätzt. Durch eine Normierung mit $1/(N-1)=1$ ergibt sich die korrekte Varianz. Auch für $N>2$ ist die Zahl der unabhängigen Residuenquadraten $N-1$, wobei das $-1$ dadurch erklärt werden kann, dass mit dem arithmetischen Mittel ja schon eine Funktion der Daten verwendet wird. Es wird dann gesagt, dass die Zahl der Freiheitsgrade (engl.: degrees of freedom) um eins reduziert wurde und nur noch $N-1$ beträgt. Das erklärt auch den Namen der Option in `np.var()`, "ddof" = "delta degrees of freedom".

Im folgenden Codebeispiel werden zehn Stichproben mit jeweils $N=10$ aus einer Standardnormalverteilung mit $\mu=0$ und $\sigma=1$ gezogen und arithmetisches Mittel, Stichprobenvarianz und Varianz ohne Besselkorrektur ausgegeben. 

```python
"""draw samples of random numbers from a normal (Gaussian) distribution, print mean/variance"""
import numpy as np
import matplotlib.pyplot as plt

# initialize random number generator with seed
rng = np.random.default_rng( 42 )

# nsamples samples of N normally distributed random numbers
nsamples = 10
N = 10
gauss = rng.normal( size=(nsamples,N) )

# print mean and variance with and without Bessel correction
print( "mean, sample variance, variance")
for mean, svar, var in zip( gauss.mean(axis=1), gauss.var(ddof=1,axis=1), gauss.var(axis=1) ):
    print( "%6.3f, %6.3f, %6.3f" % ( mean, svar, var ) )
```

#### Unsicherheit des Schätzwerts
Mit dem arithmetischen Mittel können wir einen Schätzwert für den wahren Wert einer Größe aus Messdaten berechnen. Dieser Vorgang wird in der mathematischen Statistik als **Punktschätzung** bezeichnet. In der Physik ist zusätzlich die Präzision von Interesse, mit der wir den wahren Wert schätzen können, oder anders ausgedrückt die **statistische Unsicherheit** der Messung. Dies entspricht in der Statistik einer **Intervallschätzung**. Die Stichprobenvarianz bzw. -standardabweichung ist ein Schätzwert für die Streuung der Daten, sie ist aber **nicht** ein Maß für die Unsicherheit des Schätzwerts. Diese Größe wird im folgenden motiviert.

Die Unsicherheit des Schätzwerts $\sigma_{\hat\mu}$ wird in der Literatur auch oft als **Standardunsicherheit** oder **Stichprobenunsicherheit** bezeichnet. Das Quadrat dieser Größe $\sigma_{\hat\mu}^2$ kann als Varianz der Stichprobenvarianz unter Verwendung der Rechenregeln für die Varianz berechnet werden:
$$
\sigma_{\hat\mu}^2 = \mathsf{V}[\hat\mu] 
= \frac{1}{N^2} \sum\limits_{i=1}^{N} \mathsf{V}[x_i] 
= \frac{1}{N^2} \cdot N \cdot \hat\sigma_x^2
= \frac{1}{N(N-1)} \sum\limits_{i=1}^{N} ( x_i - \hat\mu )^2.
$$
Später in diesem Kapitel wird außerdem gezeigt, wie diese Gleichung mithilfe der linearen Fehlerfortpflanzung abgeleitet werden kann. Damit ist die Standardunsicherheit um einen Faktor $1/\sqrt{N}$ kleiner als die Stichprobenstandardabweichung:
$$
\sigma_{\hat\mu}=\frac{\hat \sigma}{\sqrt{N}}.
$$
Für die Verbesserung eines wissenschaftlichen Ergebnisses durch häufige Wiederholung der Messung ist dies ein wichtiges Ergebnis: für $N$ Messungen derselben physikalischen Größe **reduziert sich die Standardunsicherheit um den Faktor $1/\sqrt{N}$** im Vergleich zu einer einzigen Messung, siehe Codebeispiel unten. Diese Tatsache hatten wir beim Beispiel der Wahrscheinlichkeit eines Münzwurfs bei der Einführung des frequentistischen Wahrscheinlichkeitsbegriffs am Anfang des Kapitels bereits verwendet. Bei physikalischen Experimenten, für die die Genauigkeit des Resultats mehr von statistischen als von systematischen Unsicherheiten abhängt, wird dieser Zusammenhang auch bei der Abschätzung der notwendigen Messzeit verwendet. Mit zehnfacher Datenmenge verbessert sich die Genauigkeit beispielsweise um einen Faktor $\sqrt{10}\approx 3.16$.

```python
"""illustration of standard uncertainty from random numbers from normal distribution"""

import numpy as np
import matplotlib.pyplot as plt

# initialize random number generator with seed
rng = np.random.default_rng( 42 )

# five samples of different size
Narray = [ 10, 100, 1000, 10000, 100000 ]

print( "sample, mean, sample standard deviation, standard uncertainty")
for N in Narray:
    gauss = rng.normal( size=N )
    print( "%6d %6.3f, %6.3f, %6.3f" % ( N, gauss.mean(), gauss.std(ddof=1), gauss.std(ddof=1)/np.sqrt(N) ) )
```

### Darstellung von Messgrößen

Mit dem arithmetischen Mittel der Messwerte als Schätzwert für eine Messgröße und der Standardunsicherheit als Maß für die Unsicherheit dieses Schätzwerts bietet es sich an, diese beiden Größen für die Darstellung von Messgrößen und ihre Unsicherheit zu verwenden:
$$
\hat\mu \pm \sigma_{\hat\mu}
$$
Dies ist in der Physik die üblichen Konvention zur Angabe von Messgrößen. Grafisch wird der Ausdruck $\hat\mu \pm \sigma_{\hat\mu}$ durch einen **Unsicherheitsbalken** (engl.: uncertainty bar, früher: Fehlerbalken, engl.: error bar) symbolisiert, mit einem Datenpunkt und einem Balken der Länge $2\sigma_{\hat\mu}$ symmetrisch um den Datenpunkt. 

>**Exkurs: Unsicherheitsbalken, Konfidenzintervall und Abdeckung**
>
> Wenn  Messdaten normalverteilt sind, deckt der Unsicherheitsbalken das frequentistische Konfidenzintervall zwischen den Quantilen $x_{0.159}$ und $x_{0.841}$ ab, für eine einzige Messung der ursprünglichen Verteilung, ansonsten um den Faktor $1/\sqrt{N}$ gestaucht. Dies deckt die zentralen 68.3% der Verteilung ab. Das bedeutet übrigens auch, dass nach frequentistischer Denkweise in 100%-68.3%=31.7% aller Fälle der der Unsicherheitsbalken nicht den wahren Wert enthält. Diese Tatsache wird gern als "sanity check" für korrekte Unsicherheitsbalken verwendet. Auch für Messdaten, die nicht normalverteilt wird, werden in der Physik Konfidenzintervalle mit 68.3% Abdeckung verwendet, diese Intervalle können dann auch asymmetrisch um den Datenpunkt angeordnet sein. In fortgeschrittenen Statistiklehrbüchern finden Sie Informationen zur **Neyman-Konstruktion** als systematische Vorschrift zur Konstruktion frequentistischer Konfidenzintervalle.

Andere Wissenschaftsbereiche stellen häufig nur gemessene Verteilungen in komprimierter Form dar, nicht Messgrößen und Unsicherheiten. Sie verfolgen dann auch andere Konventionen der Darstellung. In der deskriptiven Statistik sind **Box(-Whisker)-Plots** sehr verbreitet:

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/b/b1/Elements_of_a_boxplot.svg/354px-Elements_of_a_boxplot.svg.png">  

Hier wird für eine Verteilung dargestellt:
- der **Median** $x_{0.5}$ anstatt des arithmetischen Mittels als Schätzwert der Größe verwendet, dies ist robuster gegen Ausreißer;
- eine **Box** mit den Grenzen $x_{0.25}$ (unteres Quartil Q1) und $x_{0.75}$ (oberes Quartil Q3), die 50% Abdeckung bietet, also nur ein Bruchteil 0.675 der Abdeckung des Unsicherheitsbalkens von 68.3%;
- **Antennen** (engl.: whiskers) um die Ausläufer der Verteilung zu symbolisieren, oft mit der Konvention $Q1 - 1.5(Q3-Q1)$ nach unten und $Q3 + 1.5(Q3-Q1)$ nach oben, also zusätzlich dem 1,5-fachen des "inter-quartile range" $Q3-Q1$ nach oben und unten.
- ggf. zusätzlich einzelne Ausreißer als Kreise.

Im folgenden Codebeispiel wird für je 100 Zufallszahlen aus einer Normalverteilung und einer Exponentialverteilung ein Box-Whisker-Plot mit dem Ansatz des Unsicherheitsbalkens verglichen. Vorsicht, dabei wird der Unsicherheitsbalken neben der üblichen Darstellung der **Unsicherheit der Messgröße** (grün) auch zur Darstellung der **Streuung der Daten** verwendet (rot).

```python
"""illustrate different ways of representing distributions and uncertainties"""

import numpy as np
import matplotlib.pyplot as plt

def uncertainty_bars( data, ax ):

    # lines at the random data
    ax.vlines(data, 0, 1, linestyles='-', lw=2)

    # classic uncertainty bar symbolizing uncertainty of mean
    meanx = np.mean( data )
    meany = 2.
    stdx  = np.std( data, ddof=1 )/np.sqrt( len(data) )
    ax.errorbar( meanx, meany, xerr=stdx, fmt='-go' )

    # non-standard uncertainty bar symbolizing spread of distribution itself
    meanx = np.mean( data )
    meany = 3.
    stdx  = np.std( data, ddof=1 )
    ax.errorbar( meanx, meany, xerr=stdx, fmt='-ro' )

    # box-whisker plot
    boxy = 4.
    ax.boxplot( data, positions=[boxy], vert=False , notch=False )

    ax.set_ylim( 0, 5 )
    ax.set_yticks ( [] )
    ax.set_xlabel( "$x$") 

    return 

rng = np.random.default_rng(42)
n = 100

fig,ax = plt.subplots( 2, 1, figsize=(7,10) )

uncertainty_bars( rng.normal( size=n ), ax[0] )
uncertainty_bars( rng.exponential( size=n ), ax[1] )

plt.show()
```

>**Exkurs: Signifikanten Stellen und Angabe von Unsicherheiten bei Messergebnissen**
>
>Messergebnisse besitzen **immer** eine Unsicherheit, und die Angabe der Ergebnisse, sei es im Praktikum oder in einer wissenschaftlichen Veröffentlichung, sollte das korrekt reflektieren. Während bei der Auswertung am Computer alle Zwischenrechnungen möglichst mit der vollen numerischen Präzision und ohne zusätzliche Rundung erfolgen sollten, werden Ergebnis und Unsicherheit in der Regeln **gerundet auf höchstens zwei signifikante Stellen (engl.: significant figures) der Unsicherheit** angegeben. 
>
>Unter dem Begriff "signifikante Stellen" versteht z. B. die DIN 1333 ein Anzahl der Stellen von der ersten von Null verschiedenen Stelle bis zur letzten Stelle, die nach dem Runden noch angegeben wird ("Rundestelle"). Dabei dürfen bei Messergebnissen keine Nullen nach der letzten Stelle aufgefüllt werden. Die Rundung muss außerdem so erfolgen, dass die relative Abweichung beim Runden ("Rundungsfehler") nicht zu groß wird. Daraus ergeben sich folgende Regeln für die Rundung auf eine Stelle:
>- Suche von links beginnend und ohne Beachtung des Dezimalkommas (bzw. des Dezimalpunkts) die erste von Null verschiedene Ziffer der Messunsicherheit.
>- Ist diese Ziffer $\geq 3$, ist diese Ziffer die Rundestelle, hat sie den Wert $1$ oder $2$, ist die folgende Stelle die Rundestelle. Mit dieser Regel werden zu große relative Abweichungen beim Runden vermieden.
>- Die **Messunsicherheit** wird **immer aufgerundet**, unabhängig vom Wert der folgenden Stelle, um nicht durch die Rundung eine zu kleine Unsicherheit zu erhalten.
>- Das **Messergebnis** wird nach den bekannten Regeln **aufgerundet** (Wert der folgenden Stelle $\geq 5$) **oder abgerundet** (Wert der folgenden Stelle $<4$).
>
>**Beispiel 1**: Eine Hochspannungsmessung ergibt $\hat V=1234.5\,\mathsf{V}$. Die Unsicherheit beträgt $\sigma_{\hat{V}} = 54.321\,\mathsf{V}$. Die erste von Null verschiedene Stelle der Unsicherheit ist die Zehnerstelle, und mit einem Wert von $5$ ist diese Stelle die Rundestelle. Die Unsicherheit wird aufgerundet auf $\sigma_{\hat{V}} = 0.06\,\mathsf{kV}$. Diese Schreibweise wird bevorzugt vor $\sigma_{\hat{V}} = 60\,\mathsf{V}$ um keine Nullen auffüllen zu müssen. Entsprechend wird auch das Ergebnis an derselben Stelle gerundet, und es ergibt sich
>$$
V = 1.23 \pm 0.06 \,\mathsf{kV}.
>$$
>
>**Beispiel 2**: Eine Strommessung ergibt $\hat I=0.054321\,\mathsf{A}$ mit einer Unsicherheit von $\sigma_{\hat{I}} = 0.0012345\,\mathsf{A}$. Die ersten von Null verschiedene Stelle der Unsicherheit ist die dritte Nachkommastelle. Da die Ziffer dort $1$ lautet, wird die folgende Stelle, also die vierte Nachkommastelle, zur Rundestelle. Die Unsicherheit wird an dieser Stelle aufgerundet auf $\sigma_{\hat{I}} = 0.0013\,\mathsf{A}$. Das Ergebnis muss an derselben Stelle auf- oder abgerundet werden. Der Wert der Ziffer ist $3$, daher wird abgerundet. Das gerundete Ergebnis ist damit
>$$
I = 0.0543 \pm 0.0013 \,\mathsf{A} \quad\textsf{oder}\quad
I = 54.3 \pm 1.3 \,\mathsf{mA} 
>$$
>Eine alternative Schreibweise ist eine Angabe der Unsicherheit in Klammern nach dem Wert
>$$
I = 0.0543(13) \,\mathsf{A}.
>$$
